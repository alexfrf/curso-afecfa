Analizaremos los centros de nuestro rival -**Leeds United**- tomando como referencia el último partido disputado por los whites, en su victoria 2-1 frente al Wolverhampton Wanderers en la primera jornada de la Premier League 2022-23 

El primer paso, como siempre, es el de importar las librerías necesarias. Añadimos también la ruta del archivo de eventing que vamos a emplear.

```python
import matplotlib.pyplot as plt
import pandas as pd
from mplsoccer.pitch import Pitch
import xml.etree.ElementTree as et

from urllib.request import urlopen

import matplotlib.patheffects as path_effects
import matplotlib.pyplot as plt
import pandas as pd
from PIL import Image
import numpy as np
import os
from pathlib import Path
from mplsoccer import Pitch, FontManager, add_image
```

```python
base_dir = os.path.join('',Path(os.getcwd()).parents[1])
data_dir = os.path.join(base_dir, 'Data/Eventing_PremierLeague')
file = '06-08-2022--England--Premier-LeagueLeeds-United-2-1-Wolverhampton-Wanderers20220810-97111-12o8ztk.xml'
```

Procedemos a la importación del xml y su conversión a dataframe de pandas.

```python
file = os.path.join(data_dir,file)
tree = et.parse(file)
root = tree.getroot()

node_names = ['ID', 'start', 'end', 'code', 'pos_x', 'pos_y']
node_names_full = []
matrix = []
for node in root.iter('ALL_INSTANCES'):
    for instance in node.iter('instance'):
        if instance.find('code').text!='Empezar marca de tiempo':
            row=[]
            for node_name in node_names:
                row.append(instance.find(node_name).text)
            for label in instance.iter('label'):
                row.append(label.find('text').text)
                if label.find('group').text not in node_names_full and label.find('group').text not in node_names:
                    node_names_full.append(label.find('group').text)
            matrix.append(row)
for i in node_names_full:
    node_names.append(i)
df = pd.DataFrame(data = matrix,columns = node_names)
df.head()
```

Renombramos las columnas con nombre menos explicativo y, tras ello, transformamos las columnas numéricas, que vienen en formato no numéricos.

```python
df.rename({'code':'Player','Medio':'Half'},axis=1,inplace=True)
df.head()
```

```python
for i in ['ID','start','end','pos_x','pos_y']:
    df[i] = pd.to_numeric(df[i])
    
df.info()
```

Los datos de instat no traen una línea temporal intuitiva, sino dos variables: start y end. A partir de ellas, calculamos el minuto de juego en el que se producen las acciones.

```python
min_45 = 45*60
descanso = df[df['Half'].str.contains('1')].tail(1)['end'].values[0]
descuento_1_parte = descanso - min_45
df['start_real'] =df.start
df['start_real'] = np.where((df.start>min_45) & (df.Half.str.contains('1')),min_45,df.start_real)
df['start_real'] = np.where(df.Half.str.contains('2'),df.start-descuento_1_parte,df.start_real)
df['Min'] = round((df.start_real/60),0)
df['Min'] = np.where(df.Min==0,1,df.Min)
df['Min'] = df['Min'].astype(int)
df.drop('start_real',axis=1,inplace=True)
```

### Diagrama Pases - Marc Roca

A continuación, mostraremos el diagrama con todos los pases (tanto precisos como imprecisos) del mediocentro del Leeds United, Marc Roca, en su estreno como jugador del Leeds. El español fue sustituido a los 72 minutos de partido.

```python
instat_y = 68
instat_x = 105
df_pases = df[(df['Accion']=='Pases precisos') | (df['Accion']=='Pases (inefectivo)')]
```

A continuación, tratamos de extraer, para cada pase, información relevante que no está, en origen, incluida en las filas de las tablas. Buscamos obtener las coordenadas de destino del pase (REC_X, REC_Y), el nombre del receptor (REC_JUGADOR) y si el pase ha ido a un compañero o a un rival (REC_OUTCOME).

```python
df_pases['Rec_x'] = 0
df_pases['Rec_y'] = 0
df_pases['Rec_Jugador'] = ''
df_pases['Rec_Outcome'] = ''

for i,j in df_pases.iterrows():
    player = df_pases.loc[i,'Player']
    team = df_pases.loc[i,'Equipo']
    for k,h in df[i:].iterrows():
        if player!=h['Player']:
            df_pases.loc[i,'Rec_x']= df.loc[k,'pos_x']
            df_pases.loc[i,'Rec_y']= df.loc[k,'pos_y']
            df_pases.loc[i,'Rec_Jugador']= df.loc[k,'Player']
            if team==h['Equipo']:
                df_pases.loc[i,'Rec_Outcome']= 'Propio'
            else:
                df_pases.loc[i,'Rec_Outcome']= 'Rival'
                df_pases.loc[i,'Rec_x']= instat_x-df.loc[k,'pos_x']
                df_pases.loc[i,'Rec_y']= instat_y-df.loc[k,'pos_y']
            break
```

```python
df_pases_filt = df_pases[(df_pases['Player'].str.contains('Marc Roca'))]
df_pases_filt.head()
```

Mostramos el número de pases acertados y fallados, además de los receptores de los pases -que sean sus compañeros- y el número de pases completados recibidos por estos jugadores.

```python
df_pases_filt.Accion.value_counts()
```

```python
df_pases_filt[df_pases_filt.Rec_Outcome=='Propio'].Rec_Jugador.value_counts()
```

A continuación, mostramos el diagrama, donde los pases completados figuran en azul y los fallados en rojo. Se anota también el receptor y eñ minuto de juego.

```python
# Setup the pitch
pitch = Pitch(pitch_type='uefa', pitch_color='#386D38', line_color='#c7d5cc')
fig, ax = pitch.draw(figsize=(16, 14), constrained_layout=True, tight_layout=False)
fig.set_facecolor('#386D38')

df_pases_filt_p = df_pases_filt[df_pases_filt.Rec_Outcome=='Propio']
df_pases_filt_i = df_pases_filt[df_pases_filt.Rec_Outcome=='Rival']

sc = pitch.scatter(df_pases_filt_p.pos_x,df_pases_filt_p.pos_y,
                  color='navy',marker='o',
                  s=100,ax=ax,edgecolors='black')
pitch.arrows(df_pases_filt_p.pos_x, df_pases_filt_p.pos_y,
             df_pases_filt_p.Rec_x, df_pases_filt_p.Rec_y, width=2,
             headwidth=10, headlength=10, color='navy', ax=ax, label='Pases Precisos')
sc = pitch.scatter(df_pases_filt_i.pos_x,df_pases_filt_i.pos_y,
                  color='red',marker='o',
                  s=100,ax=ax,edgecolors='black')
pitch.arrows(df_pases_filt_i.pos_x, df_pases_filt_i.pos_y,
             df_pases_filt_i.Rec_x, df_pases_filt_i.Rec_y, width=2,
             headwidth=10, headlength=10, color='red', ax=ax, label='Pases Imprecisos')

for i in df_pases_filt.values:
    plt.text(i[10]+1,i[11]-0.5,"{}, {}´".format(i[-2],i[-5]),size=11,color='black',weight='bold')

ax.legend( handlelength=5, edgecolor='None', fontsize=15, loc='upper left')

ax.set_title('{} - Pases vs {}'.format(df_pases_filt.Player.values[0],'Wolves'), fontsize=30)
plt.show();
```

### Presiones FBRef

Mostramos a continuación los diagramas de presiones correspondientes a la temporada 2020-21 de la Premier League. Se emplea esta temporada, y no la actual o la 2021-2022 porque, a fecha de realización de este ejercicio, los datos de presiones ya no están disponibles en FbRef.

```python
%matplotlib inline
```

En primer lugar, tomamos, directamente desde las tablas alojadas en la web, los datos de los equipos de Premier League de aquel año, simplemente para poder extraer una lista con los equipos que formaban parte de esa liga durante esa temporada, y poder filtrar nuestro dataframe (que contiene los equipos de las cinco principales ligas).

```python
URL = 'https://fbref.com/en/comps/9/2020-2021/defense/2020-2021-Premier-League-Stats#stats_squads_defense_for'
equipos_ing = pd.read_html(URL)[0]
equipos_ing.columns = equipos_ing.columns.droplevel()
equipos_ing = equipos_ing.Squad.unique()
```

Importamos el dataframe desde excel y nos quedamos solo con los equipos ingleses.

```python
df = pd.read_excel(data_dir+'/fbref_data.xlsx')
df = df[df.Team.isin(equipos_ing)]
df.shape[0]
```

Comprobamos que todas las columnas tienen un formato (tipo) correcto.

```python
df.info()
```

Almacenamos en df_total la media, para cada tercio, de las presiones realizadas por los equipos de Premier League en esa temporada. Dicho parámetro nos servirá como referencia para la visualización adicional que veremos abajo.

```python
pressure_cols = ['Press Def 3rd', 'Press Mid 3rd', 'Press Att 3rd']
df_total = pd.DataFrame(df[pressure_cols].mean())
df_total.columns = ['total']
df_total = df_total.T
#df_total = df_total.divide(df_total.sum(axis=1), axis=0) * 100
```

```python
pcts = df.copy()
p_dif = df.copy()
pcts[pressure_cols] = pcts[pressure_cols].divide(pcts[pressure_cols].sum(axis=1), axis=0) * 100.
```

Establecemos efectos visuales y tomamos el logo de Statsbomb para los diagramas de dispersión a continuación.

```python
LOGO_URL = 'https://github.com/statsbomb/open-data/blob/master/img/SB%20-%20Icon%20Lockup%20-%20Colour%20positive.png?raw=true'
sb_logo = Image.open(urlopen(LOGO_URL))

fm = FontManager()
path_eff = [path_effects.Stroke(linewidth=3, foreground='black'),
            path_effects.Normal()]
```

De la url de FBREF podemos extraer el nombre de la liga, con un par de funciones asociadas al manejo de strings.

```python
url1 = URL.split('defense/')[1]
url1 = url1.split('-Stats')[0]
seas = url1[:9]
comp =url1[10:].replace('-',' ')
```

Realizamos las dos visualizaciones. En la primera, mostramos el porcentaje de presiones por tercio de campo que ha hecho cada equipo. Dicho diagrama muestra en qué punto inciden su presión e intensidad defensiva cada equipo. Sin embargo, no es suficiente para aclarar otros puntos clave, como la vocación defensiva de cada equipo. Es decir, no aclara si un equipo es particularmente intenso o no, sino exclusivamente donde se da mayor o menor intensidad.

Para solventar lo comentado surge la segunda visualización. En ella, podemos ver el diferencial, para cada zona, de presiones realizadas con respecto a la media del campeonato. Un % positivo implica que un equipo ha presionado, en esa zona, un x% más que la media del campeonato.

```python
pitch = Pitch(line_zorder=2, line_color='black', pad_top=20)

bin_statistic = pitch.bin_statistic([0], [0], statistic='count', bins=(3, 1))

GRID_HEIGHT = 0.8
CBAR_WIDTH = 0.03
fig, axs = pitch.grid(nrows=4, ncols=5, figheight=20,
                      # leaves some space on the right hand side for the colorbar
                      grid_width=0.88, left=0.025,
                      endnote_height=0.06, endnote_space=0,
                      axis=False,
                      title_space=0.02, title_height=0.06, grid_height=GRID_HEIGHT)
fig.set_facecolor('white')

teams = df['Team'].values
vmin = pcts[pressure_cols].min().min()  
vmax = pcts[pressure_cols].max().max()
for i, ax in enumerate(axs['pitch'].flat[:len(teams)]):
    ax.text(60, -10, teams[i],
            ha='center', va='center', fontsize=35,
            fontproperties=fm.prop)

    bin_statistic['statistic'] = pcts.loc[pcts.Team == teams[i], pressure_cols].values
    heatmap = pitch.heatmap(bin_statistic, ax=ax, cmap='coolwarm', vmin=vmin, vmax=vmax)
    annotate = pitch.label_heatmap(bin_statistic, color='white', fontproperties=fm.prop,
                                   path_effects=path_eff, fontsize=45, ax=ax,
                                   str_format='{0:.0f}%', ha='center', va='center')

if len(teams) == 18:
    for ax in axs['pitch'][-1, 3:]:
        ax.remove()

cbar_bottom = axs['pitch'][-1, 0].get_position().y0
cbar_left = axs['pitch'][0, -1].get_position().x1 + 0.01
ax_cbar = fig.add_axes((cbar_left, cbar_bottom, CBAR_WIDTH,
                        GRID_HEIGHT - 0.036))
cbar = plt.colorbar(heatmap, cax=ax_cbar)
for label in cbar.ax.get_yticklabels():
    label.set_fontproperties(fm.prop)
    label.set_fontsize(20)

# title and endnote
add_image(sb_logo, fig,
          left=axs['endnote'].get_position().x0,
          bottom=axs['endnote'].get_position().y0,
          height=axs['endnote'].get_position().height)
title = axs['title'].text(0.5, 0.5, 'Presiones %, {}, {}'.format(comp,seas),
                          ha='center', va='center', fontsize=60)
```

Como comentábamos, este diagrama nos permite saber qué equipos prefieren armar bloques medios/bajos y defender exclusivamente en el espacio del campo propio (caso de Wolves, West Brom o Tottenham) y aquellos que presentan un mayor balance entre las presiones realizadas. Si exclusivamente nos ceñimos a esta visualización, sin conocer nada más sobre los equipos, no sabríamos si equipos como Liverpool o Manchester City son muy presionantes arriba o, simplemente, no son intensos en ninguna zona del campo (al fin y al cabo hablamos de porcentajes, no podemos conocer el volumen de presiones realizadas en cada zona ni en total observando a este gráfico). Además, se da otra particularidad: midiendo los % de cada tercio sobre el total, los equipos que son muy agresivos en la presión en todas las zonas del campo por igual pasan desapercibidos -caso del Leeds de Marcelo Bielsa, que sí destaca en la visualización que vemos abajo-.

```python
p_dif[pressure_cols] =  100 * (p_dif[pressure_cols].values - df_total.values) / df_total.values

GRID_HEIGHT = 0.76
fig, axs = pitch.grid(nrows=4, ncols=5, figheight=20,
                      grid_width=0.88, left=0.025,
                      endnote_height=0.06, endnote_space=0,
                      axis=False,
                      title_space=0.02, title_height=0.1, grid_height=GRID_HEIGHT)
fig.set_facecolor('white')

teams = p_dif['Team'].values
vmin = p_dif[pressure_cols].min().min() 
vmax = p_dif[pressure_cols].max().max()

for i, ax in enumerate(axs['pitch'].flat[:len(teams)]):
    ax.text(60, -10, teams[i], ha='center', va='center', fontsize=30, fontproperties=fm.prop)

    bin_statistic['statistic'] = p_dif.loc[p_dif.Team == teams[i], pressure_cols].values
    heatmap = pitch.heatmap(bin_statistic, ax=ax, cmap='coolwarm', vmin=vmin, vmax=vmax)
    annotate = pitch.label_heatmap(bin_statistic, color='white', fontproperties=fm.prop,
                                   path_effects=path_eff, str_format='{0:.0f}%', fontsize=45,
                                   ax=ax, ha='center', va='center')

if len(teams) == 18:
    for ax in axs['pitch'][-1, 3:]:
        ax.remove()

cbar_bottom = axs['pitch'][-1, 0].get_position().y0
cbar_left = axs['pitch'][0, -1].get_position().x1 + 0.01
ax_cbar = fig.add_axes((cbar_left, cbar_bottom, CBAR_WIDTH,
                        # take a little bit off the height because of padding
                        GRID_HEIGHT - 0.035))
cbar = plt.colorbar(heatmap, cax=ax_cbar)
for label in cbar.ax.get_yticklabels():
    label.set_fontproperties(fm.prop)
    label.set_fontsize(25)

add_image(sb_logo, fig,
          left=axs['endnote'].get_position().x0,
          bottom=axs['endnote'].get_position().y0,
          height=axs['endnote'].get_position().height)
TITLE = 'Presiones, Diferencia Porcentual Respecto al Promedio por Tercio\n{} {}'.format(comp,seas)
title = axs['title'].text(0.5, 0.5, TITLE, ha='center', va='center', fontsize=45)
plt.show();
```
