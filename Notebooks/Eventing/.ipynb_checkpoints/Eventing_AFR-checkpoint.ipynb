```python
import matplotlib.pyplot as plt
import pandas as pd
from mplsoccer.pitch import Pitch
import xml.etree.ElementTree as et
import os
from pathlib import Path
```

```python
base_dir = os.path.join('',Path(os.getcwd()).parents[1])
data_dir = os.path.join(base_dir, 'Data/EventingEibar')
file = '07-02-2021--Spain--Primera-DivisionOsasuna-2-1-Eibar20210306-3162-1isaph3.xml'
```

```python
file = os.path.join(data_dir,file)

tree = et.parse(file)
root = tree.getroot()

node_names = ['ID', 'start', 'end', 'code', 'pos_x', 'pos_y']
node_names_full = []
matrix = []
for node in root.iter('ALL_INSTANCES'):
    for instance in node.iter('instance'):
        if instance.find('code').text!='Empezar marca de tiempo':
            row=[]
            for node_name in node_names:
                row.append(instance.find(node_name).text)
            for label in instance.iter('label'):
                row.append(label.find('text').text)
                if label.find('group').text not in node_names_full and label.find('group').text not in node_names:
                    node_names_full.append(label.find('group').text)
            matrix.append(row)
```

```python
df = pd.DataFrame(data = matrix,columns = node_names+node_names_full)
df.head()
```

```python
df.rename({'code':'Player','Medio':'Half'},axis=1,inplace=True)
df.head()
```

```python
for i in ['ID','start','end','pos_x','pos_y']:
    df[i] = pd.to_numeric(df[i])
    
df.info()
```

```python
df = df.sort_values('end').reset_index().drop('index',axis=1)
df
```

### Passing Map - Bryan Gil

```python
df.head(30)
```

```python
instat_x = 105
instat_y = 68
```

```python
df_pases = df[df['Accion'].str.contains('Pases precisos') | df['Accion'].str.contains('Asistencias')]
df_pases.head()
```

```python
def detalle_jugadas(df):

    df['Rec_x'] = 0
    df['Rec_y'] = 0
    df['Pase_Receptor'] = ''
    df['Resultado_Accion'] = ''

    for i,j in df[(df['Accion'].str.contains('Pases precisos')) | (df['Accion'].str.contains('Centros'))].iterrows():
        player = df.loc[i,'Player']
        team = df.loc[i,'Equipo']

        for k,h in df[i:].iterrows():
            if player!=h['Player'] : #and h['Accion']=='Ataques posicionales':
                df.loc[i,'Pase_Receptor']= h['Player']
                if team==h['Equipo']:
                    df.loc[i,'Rec_x']= h['pos_x']
                    df.loc[i,'Rec_y']= h['pos_y']
                    df.loc[i,'Resultado_Accion']= 'Completado'
                else:
                    df.loc[i,'Resultado_Accion']= 'No Completado'
                    df.loc[i,'Rec_x']= instat_x-h['pos_x']
                    df.loc[i,'Rec_y']= instat_y-h['pos_y']
                break

    for i,j in df[(df['Accion'].str.contains('Asistencias'))].iterrows():
        player = df.loc[i,'Player']
        team = df.loc[i,'Equipo']
        for k,h in df[i:].iterrows():        
            if player!=h['Player'] and team==h['Equipo'] and h['Accion']=='Goles':
                df.loc[i,'Rec_x']= h['pos_x']
                df.loc[i,'Rec_y']= h['pos_y']
                df.loc[i,'Resultado_Accion']= 'Completado'
                df.loc[i,'Pase_Receptor']= h['Player']
                break

    for i,j in df[(df['Accion'].str.contains('Conduccion'))].iterrows():
        player = df.loc[i,'Player']
        for k,h in df[i:].iterrows():
            if player==h['Player'] and (j['pos_x']!=h['pos_x'] or j['pos_y']!=h['pos_y']):
                df.loc[i,'Rec_x']= h['pos_x']
                df.loc[i,'Rec_y']= h['pos_y']
                df.loc[i,'Resultado_Accion'] = h['Accion']

                break
    return df

df = detalle_jugadas(df)
```

```python
df_pases = df[(df['Accion'].str.contains('Pases precisos')) | (df['Accion'].str.contains('Asistencias')) | (df['Accion'].str.contains('Centros')) ]
```

```python
df_pases.info()
```

```python
df_pases_filt = df_pases[(df_pases['Equipo']=='Eibar') & (df_pases['Player'].str.contains('Bryan Gil'))]
df_pases_filt.head()
```

```python
# Setup the pitch
pitch = Pitch(pitch_type='uefa', pitch_color='#386D38', line_color='#c7d5cc')
fig, ax = pitch.draw(figsize=(16, 11), constrained_layout=True, tight_layout=False)
fig.set_facecolor('#386D38')

# Plot the completed passes
df_pases_fail = df_pases_filt[df_pases_filt.Resultado_Accion=='No Completado']
pitch.arrows(df_pases_fail.pos_x, df_pases_fail.pos_y,
             df_pases_fail.Rec_x, df_pases_fail.Rec_y, width=2,
             headwidth=10, headlength=10, color='red', ax=ax, label='uncompleted passes')

df_pases_ok = df_pases_filt[df_pases_filt.Resultado_Accion=='Completado']

pitch.arrows(df_pases_ok.pos_x, df_pases_ok.pos_y,
             df_pases_ok.Rec_x, df_pases_ok.Rec_y, width=2,
             headwidth=10, headlength=10, color='blue', ax=ax, label='completed passes')

# setup the legend
ax.legend( handlelength=5, edgecolor='None', fontsize=15, loc='upper left')

# Set the title
ax.set_title('{} passes vs {}'.format(df_pases_filt.Player.values[0],'Osasuna'), fontsize=30)
plt.show();
```
