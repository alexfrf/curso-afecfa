En el presente notebook compararemos, a partir de datos de Instat, los números ofensivos de Gabriel Jesus en su última temporada en el Manchester City con los de su homólogo en la delantera del Arsenal, Alexandre Lacazette, para tratar de plasmar los puntos en los que el delantero brasileño puede mejorar al equipo, ante su reciente fichaje por los Gunners.

En primer lugar, procedemos a importar las librerías necesarias, además del excel que incluye los datos exportados de instat, que en unas pocas líneas de código limpiamos.

```python
import pandas as pd
from soccerplots.radar_chart import Radar
import warnings
import os
from pathlib import Path
warnings.filterwarnings('ignore')
```

```python
base_dir = os.path.join('',Path(os.getcwd()).parents[1])
data_dir = os.path.join(base_dir, 'Data/Comps')
img_dir = os.path.join(base_dir, 'img/Radarcharts')
df = pd.read_excel(data_dir+'/PremierLeague_2122.xlsx')
df.head()
```

```python
df = df.drop('Unnamed: 0',axis=1)
df.rename({'Unnamed: 1':'Jugador'},axis=1,inplace=True)
df = df.set_index('Jugador')
df.head()
```

Establecemos un filtro que os ayudará a descartar jugadores del dataframe. Establecemos un mínimo de encuentos disputados, que es igual a la media del campeonato - la desviación típica.

```python
min_games= int(df['Partidos jugados'].mean() - df['Partidos jugados'].std())
min_games
```

```python
df.shape[0]
```

```python
df = df[df['Partidos jugados']>=min_games]
df.shape[0]
```

Nos quedamos con las variables que nos interesa comparar, enumeradas abajo.

```python
variables = [
    'Goles',
 'Asistencias',
 'Tiros a portería',
 'Pases',
 '% de efectividad de pases',
 'Pases de finalización',
 '% de efectividad de los centros',
 'Recuperaciones en campo rival',
 'Goles esperados',
 'Disputas',
 '% disputas ganadas',
 '% disputas por arriba ganadas',
 'Regates'
]

df_vars = df[variables]
df_vars.head()
```

De cara a poder transformarlas en campos numéricos, debemos eliminar una serie de caracteres que no son compatibles con un número.

```python
df_vars = df_vars.replace('%', '', regex = True)
df_vars = df_vars.replace('-', '0', regex = True)
df_vars.head(10)
```

```python
df_vars = df_vars.apply(pd.to_numeric)
df_vars.head()
```

```python
df_vars.info()
```

Establecemos la función que emplearemos para visualizar los radares. Más abajo,elegimos a los jugadores y nos aseguramos de que aparecen en el dataframe.

En la función se añade un parámetro de entrada (comparador_tipo) que puede tomar el valor *relative* o *absolute*. En relative, el valor máximo de cada variable será = al valor máximo teniendo en cuenta solo a los dos jugadores que se comparan. En caso contrario, el valor máximo será = al máximo de la liga, lo que da lugar a radares más pequeños, pero que nos permiten comprobar el gap entre los jugadores que comparamos y las referencias del campeonato en cada métrica.

```python
def radar_sb(data,jugadores_comp,comparador_tipo='relative',pie='@alexfr'):
    df = data[data.index.isin(jugadores_comp)]
    if comparador_tipo=='relative':
        comparador = df.apply(lambda x: pd.Series([0,x.max()])).T.values.tolist()
    else:
        comparador= data.apply(lambda x: pd.Series([x.min(),x.max()])).T.values.tolist()

    title = dict(
        title_name="Radar",
        subtitle_name=df.index[0],
        subtitle_color='#B6282F',
        subtitle_name_2=df.index[-1],
        subtitle_color_2='#344D94',
        title_fontsize=18,
        subtitle_fontsize=15,
    )


    endnote = "{}\nVisualization made by: Anmol Durgapal(@slothfulwave612)\nAll units are in per90".format(pie)

    radar = Radar()
    fig,ax = radar.plot_radar(ranges = comparador,params=list(df.columns),values=df.values,
                             radar_color = [title["subtitle_color"],title["subtitle_color_2"]],
                             title = title, endnote=endnote,
                             filename='{}vs{}.jpg'.format(title["subtitle_name"].split(' ')[-1],title["subtitle_name_2"].split(' ')[-1]),
                             dpi=500,compare=True)
```

```python
jugadores = ['Gabriel Jesus','Alexandre Lacazette']
for i in jugadores:
    if i in list(df_vars.index):
        print('OK')
```

```python
radar_sb(df_vars,jugadores)
```

### Puntos adicionales: modifica los colores del radarchart

Cambiamos los colores, otorgando a cada jugador los colores del equipo al que representaban en la temporada. Además, hacemos el mismo plotting con valores de referencia absolutos.

```python
def radar_sb(data,jugadores_comp,comparador_tipo='relative',pie='@alexfr'):
    df = data[data.index.isin(jugadores_comp)]
    if comparador_tipo=='relative':
        comparador = df.apply(lambda x: pd.Series([0,x.max()])).T.values.tolist()
    else:
        comparador= data.apply(lambda x: pd.Series([x.min(),x.max()])).T.values.tolist()

    color_city = '#8AD9FC'
    color_arsenal= '#D30000'
    title = dict(
        title_name="Radar",
        subtitle_name=df.index[0],
        subtitle_color=color_city,
        subtitle_name_2=df.index[-1],
        subtitle_color_2=color_arsenal,
        title_fontsize=18,
        subtitle_fontsize=15,
    )


    endnote = "{}\nVisualization made by: Anmol Durgapal(@slothfulwave612)\nAll units are in per90".format(pie)

    radar = Radar()
    fig,ax = radar.plot_radar(ranges = comparador,params=list(df.columns),values=df.values,
                             radar_color = [title["subtitle_color"],title["subtitle_color_2"]],
                             title = title, endnote=endnote,
                             filename=img_dir+'/{}vs{}.jpg'.format(title["subtitle_name"].split(' ')[-1],title["subtitle_name_2"].split(' ')[-1]),
                             dpi=500,compare=True)
```

```python
radar_sb(df_vars,jugadores)
```
